```python
# dataframe으로 만들기
X_df = pd.DataFrame(X, columns=data.feature_names)
# column 별 상관관계
cor = X_df.corr()
plt.figure(figsize=(12,12))
# heatmap
sns.heatmap(cor, cmap='GnBu')
t = plt.xticks(fontsize=12)
t = plt.yticks(fontsize=12)
```
